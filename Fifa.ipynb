Need to convert height from feet and inches to inches

- Function to convert height from feet and inches to inches

```python
def convert_height_to_inches(height):
    if isinstance(height, str):  
        feet, inches = height.split("'")  
        inches = inches.replace('"', '')  
        total_inches = int(feet) * 12 + int(inches)  
        return total_inches
    else:
        return height
```

- Apply the height conversion function

```python
fifa['Height'] = fifa['Height'].apply(convert_height_to_inches)
fifa['Height']
```

- Function to convert monetary values to float

```python
def value_to_float(value):
    
    if isinstance(value, str):
        if 'M' in value:
            return float(value.replace('€', '').replace('M', '')) * 1e6
        elif 'K' in value:
            return float(value.replace('€', '').replace('K', '')) * 1e3
        else:
            return float(value.replace('€', ''))
    
    return value
```

```python
fifa['Wage'] = fifa['Wage'].apply(value_to_float)
fifa['Value'] = fifa['Value'].apply(value_to_float)
fifa['Release Clause'] = fifa['Release Clause'].apply(value_to_float)
```
